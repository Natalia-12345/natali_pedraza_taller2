#### 1. Abstraer ¿qué?, ¿por qué? y ¿cómo? 

     ¿Qué?
     
En este caso particular el ¿Qué? aplica para las dos visualizaciones realizadas dado que se realizaron con los mismos datos.

La base tiene un total de 62 nodos, los cuales representan a los 62 delfines estudiados, y 159 enlaces que representan las asociaciones entre estos durante el tiempo del estudio.
que
La base de datos es un data set geométrico, en este caso particular es de network con nodos y enlaces. El tipo de datos son los items, que están representados por los nodos, los cuales a su vez representan a cada uno de los 62 delfines del estudio. este es un data set estático debido a que muestra un momento especifico en el tiempo. 

    ¿Por qué?
    
La tarea de esta visualización es exponer de manera gráfica las relaciones de los delfines del experimento. Por lo tanto, las acciones de esta son presentar, explorar la topología y buscar las relaciones entre los delfines, asi como anotar a estos sujetos y derivar sus interacciones por medio de los enlaces. Para los objetivos tenemos un network data con topología de red. No fue necesario hacer transformaciones en los datos.


    ¿Cómo?

El modismo de esta visualización es una red dirigida, ya que las relaciones no son complementarias como se puede ver en la matriz adyacente presentada más adelante, la cual no es simétrica. La visualización tiene un nivel de interactividad, en el cual se puede ampliar la visualización para buscar un nodo en específico y de igual manera este nodo se puede seleccionar para ver, por medio de una saturación de color más alta, los enlaces que este nodo tiene con otros. 

#### 2. Marcas y canales

Para esta visualización tenemos una marca de puntos para los nodos y una marca de líneas para los enlaces. La posición espacial no está codificada directamente en este caso, solo se usa la codificación donde se ocasionen menos cruces y por lo tanto disminuya la oclusión. 

#### 3. ¿Cumple las reglas generales?

En este caso no aplica la regla del 3D injustificado. En el caso de 2D injustificado, en esta visualización se justifica el uso del 2D para presentar más efectivamente los enlaces entre los delfines comparándola con una presentación de una lista o tabla. Con el 2D podemos mostrar información como los nombres de los individuos y sus relaciones en un espacio más pequeño y comprensible. 

Esta visualización no depende implícitamente de la memoria y tampoco utiliza una inmersión en la que se deba tener en cuenta la resolución. No requiere capacidad de respuesta y es efectivo con respecto a la tarea con la que se realizó. 

#### 4. Mejoras posibles para la visualización 

Esta visualización tiene algunos problemas de oclusión si se ve de lejos, lo cual se puede modificar con una codificación diferente. Adicionalmente, si los datos tuvieran un dato adicional con el cual poder agruparlos, se podría hacer uso de un canal de color para diferenciarlos por categorías en la visualización. Aunque la visualización es de red dirigida con la interactividad esto no se puede ver, eso es un punto a mejorar.
